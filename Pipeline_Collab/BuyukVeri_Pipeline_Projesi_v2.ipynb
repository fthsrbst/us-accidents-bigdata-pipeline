## ðŸ§¹ BÃ¶lÃ¼m 4: Veri Temizleme ve Ã–n Ä°ÅŸleme (PySpark)

```python
# Eksik deÄŸer analizi (OPTÄ°MÄ°ZE EDÄ°LMÄ°Åž - tek seferde hesaplama)
print("ðŸ“Š Eksik DeÄŸer Analizi:")
print("="*60)

# Toplam satÄ±r sayÄ±sÄ±
total_rows = df_spark.count()
print(f"ðŸ“Š Toplam satÄ±r: {total_rows:,}")

# TÃœM sÃ¼tunlarÄ±n null deÄŸerlerini TEK SEFERDE hesapla (Ã§ok daha hÄ±zlÄ±!)
from pyspark.sql.functions import sum as spark_sum, when, col, isnan

# Her sÃ¼tun iÃ§in null sayÄ±sÄ±nÄ± hesapla
null_counts_expr = []
for column in df_spark.columns:
    # isNull kontrolÃ¼ (tÃ¼m tipler iÃ§in)
    null_counts_expr.append(
        spark_sum(when(col(column).isNull(), 1).otherwise(0)).alias(column)
    )

# Tek seferde tÃ¼m null'larÄ± say
null_counts_row = df_spark.select(null_counts_expr).collect()[0]

# SonuÃ§larÄ± DataFrame'e Ã§evir
null_counts = []
for column in df_spark.columns:
    null_count = null_counts_row[column]
    if null_count > 0:
        null_percent = (null_count / total_rows) * 100
        null_counts.append((column, null_count, null_percent))

# Eksik deÄŸerleri gÃ¶ster
null_df = pd.DataFrame(null_counts, columns=['SÃ¼tun', 'Eksik SayÄ±sÄ±', 'Eksik %'])
null_df = null_df.sort_values('Eksik %', ascending=False)
print(f"\nðŸ“‹ Eksik deÄŸer iÃ§eren {len(null_df)} sÃ¼tun bulundu:")
print(null_df.head(20).to_string(index=False))
```

```python
# Analiz iÃ§in Ã¶nemli sÃ¼tunlarÄ± seÃ§
# SÄ±nÄ±flandÄ±rma hedefi: Severity (1-4)
# KÃ¼meleme Ã¶zellikleri: CoÄŸrafi ve Ã§evresel faktÃ¶rler

selected_columns = [
    'Severity',           # Hedef deÄŸiÅŸken (1-4)
    'Start_Lat',          # Enlem
    'Start_Lng',          # Boylam
    'Distance(mi)',       # Etkilenen mesafe
    'Temperature(F)',     # SÄ±caklÄ±k
    'Humidity(%)',        # Nem
    'Pressure(in)',       # BasÄ±nÃ§
    'Visibility(mi)',     # GÃ¶rÃ¼ÅŸ mesafesi
    'Wind_Speed(mph)',    # RÃ¼zgar hÄ±zÄ±
    'Precipitation(in)',  # YaÄŸÄ±ÅŸ
    'Weather_Condition',  # Hava durumu
    'Sunrise_Sunset',     # GÃ¼ndÃ¼z/Gece
    'Civil_Twilight',     # AlacakaranlÄ±k
    'State'               # Eyalet
]

# SeÃ§ili sÃ¼tunlarla yeni DataFrame oluÅŸtur
df_selected = df_spark.select(selected_columns)
print(f"âœ… {len(selected_columns)} sÃ¼tun seÃ§ildi.")
df_selected.show(5)
```

```python
# Veri temizleme iÅŸlemleri
print("ðŸ§¹ Veri Temizleme BaÅŸlÄ±yor...")
print("="*60)

# 1. Eksik deÄŸerleri temizle (sadece sayÄ±sal sÃ¼tunlar iÃ§in ortalama ile doldur)
numeric_columns = ['Start_Lat', 'Start_Lng', 'Distance(mi)', 'Temperature(F)',
                   'Humidity(%)', 'Pressure(in)', 'Visibility(mi)',
                   'Wind_Speed(mph)', 'Precipitation(in)']

# Ortalama deÄŸerleri hesapla ve doldur
df_cleaned = df_selected

for col_name in numeric_columns:
    mean_val = df_cleaned.select(mean(col(col_name))).collect()[0][0]
    if mean_val is not None:
        df_cleaned = df_cleaned.fillna({col_name: mean_val})

# 2. Kategorik deÄŸiÅŸkenlerdeki eksik deÄŸerleri 'Unknown' ile doldur
categorical_columns = ['Weather_Condition', 'Sunrise_Sunset', 'Civil_Twilight', 'State']
for col_name in categorical_columns:
    df_cleaned = df_cleaned.fillna({col_name: 'Unknown'})

# 3. Severity deÄŸeri null olan satÄ±rlarÄ± kaldÄ±r (hedef deÄŸiÅŸken)
df_cleaned = df_cleaned.filter(col('Severity').isNotNull())

# Severity'yi gÃ¼venli biÃ§imde integer'a Ã§evir ve aralÄ±k dÄ±ÅŸÄ± deÄŸerleri temizle
df_cleaned = df_cleaned.withColumn('Severity', col('Severity').cast(IntegerType()))
df_cleaned = df_cleaned.filter(col('Severity').isin([1, 2, 3, 4]))

print(f"âœ… Temizleme sonrasÄ± kayÄ±t sayÄ±sÄ±: {df_cleaned.count():,}")
```

```python
# Kategorik deÄŸiÅŸkenleri sayÄ±sallaÅŸtÄ±r
print("ðŸ”„ Kategorik deÄŸiÅŸkenler sayÄ±sallaÅŸtÄ±rÄ±lÄ±yor...")

# StringIndexer ile kategorik deÄŸiÅŸkenleri encode et
indexers = []
for col_name in categorical_columns:
    indexer = StringIndexer(inputCol=col_name, outputCol=f"{col_name}_Index", handleInvalid="keep")
    indexers.append(indexer)

# Pipeline oluÅŸtur ve uygula
indexer_pipeline = Pipeline(stages=indexers)
df_indexed = indexer_pipeline.fit(df_cleaned).transform(df_cleaned)

print("âœ… Kategorik deÄŸiÅŸkenler encode edildi!")
df_indexed.select([f"{c}_Index" for c in categorical_columns]).show(5)
```

```python
# ============================================================================
# TÃœM VERÄ°YÄ° KULLAN - Ã–RNEKLEME YOK
# ============================================================================
# PySpark daÄŸÄ±tÄ±k iÅŸlem yapacaÄŸÄ± iÃ§in tÃ¼m veriyi kullanabiliriz
# Sadece Pandas'a Ã§evirirken parÃ§alÄ± iÅŸleme yapacaÄŸÄ±z

print("ðŸ“Š TÃœM VERÄ° kullanÄ±lacak - Ã¶rnekleme yapÄ±lmÄ±yor!")
print("="*60)

# TÃ¼m veriyi df_sample olarak ata (isim uyumluluÄŸu iÃ§in)
df_sample = df_indexed

# Veriyi cache'le (tekrar tekrar okumamak iÃ§in)
df_sample.cache()

total_count = df_sample.count()
print(f"âœ… Toplam veri boyutu: {total_count:,} kayÄ±t")
print(f"ðŸ“Š Bu veri K-Means ve JSON export iÃ§in kullanÄ±lacak")
print(f"ðŸ“Š kNN iÃ§in stratified sampling yapÄ±lacak (2M kayÄ±t)")
```

## ðŸŒ² BÃ¶lÃ¼m 5: Random Forest (7M+ KayÄ±t)

kNN'yi Ã§alÄ±ÅŸtÄ±rmadan Ã¶nce, aynÄ± Ã¶zelliklerle PySpark Random Forest modeli kurarak
7 milyondan fazla kaydÄ± doÄŸrudan Spark Ã¼zerinde eÄŸitiyoruz.


```python
# ============================================================================
# ðŸŒ² RANDOM FOREST SINIFLANDIRMA - PySpark MLlib
# ============================================================================
# Bu hÃ¼cre kaza ÅŸiddetini (Severity 1-4) tahmin etmek iÃ§in
# PySpark Random Forest kullanÄ±r. BÃ¼yÃ¼k veri iÃ§in optimize edilmiÅŸtir.

print("ðŸŒ² Random Forest EÄŸitimi BaÅŸlÄ±yor")
print("="*60)

from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.storagelevel import StorageLevel
import time

# Ã–zellik sÃ¼tunlarÄ± tanÄ±mla
rf_feature_columns = [
    'Start_Lat', 'Start_Lng', 'Distance(mi)',
    'Temperature(F)', 'Humidity(%)', 'Pressure(in)',
    'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)',
    'Weather_Condition_Index', 'Sunrise_Sunset_Index',
    'Civil_Twilight_Index', 'State_Index'
]

print(f"ðŸ“Š KullanÄ±lacak Ã¶zellik sayÄ±sÄ±: {len(rf_feature_columns)}")
print(f"ðŸ“Š Ã–zellikler: {', '.join(rf_feature_columns[:5])}...")
```

```python
# ============================================================================
# VERÄ° HAZIRLAMA
# ============================================================================

# Eksik deÄŸerleri temizle
df_rf_clean = df_sample
for col_name in rf_feature_columns:
    df_rf_clean = df_rf_clean.filter(col(col_name).isNotNull())

# Severity'nin geÃ§erli olduÄŸundan emin ol (1-4 arasÄ±)
df_rf_clean = df_rf_clean.filter(col('Severity').isin([1, 2, 3, 4]))

# Ã–zellik vektÃ¶rÃ¼ oluÅŸtur
assembler = VectorAssembler(
    inputCols=rf_feature_columns,
    outputCol="features",
    handleInvalid="skip"
)

# Ã–zellik Ã¶lÃ§ekleme (StandardScaler)
scaler = StandardScaler(
    inputCol="features",
    outputCol="scaled_features",
    withStd=True,
    withMean=False
)

# Pipeline uygula
df_assembled = assembler.transform(df_rf_clean)
scaler_model = scaler.fit(df_assembled)
df_rf_prepared = scaler_model.transform(df_assembled).select('scaled_features', 'Severity')
df_rf_prepared = df_rf_prepared.withColumnRenamed('scaled_features', 'features')

# Partition optimizasyonu
num_partitions = max(200, spark.sparkContext.defaultParallelism * 2)
df_rf_prepared = df_rf_prepared.repartition(num_partitions)
df_rf_prepared = df_rf_prepared.persist(StorageLevel.MEMORY_AND_DISK)

total_count = df_rf_prepared.count()
print(f"âœ… HazÄ±rlanan veri sayÄ±sÄ±: {total_count:,}")

# SÄ±nÄ±f daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶ster
print("\nðŸ“Š Severity SÄ±nÄ±f DaÄŸÄ±lÄ±mÄ±:")
df_rf_prepared.groupBy('Severity').count().orderBy('Severity').show()
```

```python
# ============================================================================
# EÄžÄ°TÄ°M / TEST BÃ–LME
# ============================================================================

# Stratified sampling benzeri etki iÃ§in seed kullan
train_rf, test_rf = df_rf_prepared.randomSplit([0.8, 0.2], seed=42)
train_rf = train_rf.persist(StorageLevel.MEMORY_AND_DISK)
test_rf = test_rf.persist(StorageLevel.MEMORY_AND_DISK)

train_count = train_rf.count()
test_count = test_rf.count()

print(f"ðŸ“Š EÄŸitim seti: {train_count:,} kayÄ±t (%80)")
print(f"ðŸ“Š Test seti: {test_count:,} kayÄ±t (%20)")
```

```python
# ============================================================================
# RANDOM FOREST MODEL EÄžÄ°TÄ°MÄ° (BELLEK DOSTU)
# ============================================================================

print("ðŸŒ² Random Forest modeli eÄŸitiliyor...")
print("="*60)

# Model parametreleri - BELLEK DOSTU ayarlar
rf_classifier = RandomForestClassifier(
    labelCol='Severity',
    featuresCol='features',
    numTrees=50,                # AÄŸaÃ§ sayÄ±sÄ± (bellek iÃ§in azaltÄ±ldÄ±)
    maxDepth=8,                 # Maksimum derinlik (bellek iÃ§in azaltÄ±ldÄ±)
    maxBins=64,                 # Maksimum bin sayÄ±sÄ±
    minInstancesPerNode=20,     # Yapraktaki minimum Ã¶rnek
    subsamplingRate=0.7,        # Bootstrap Ã¶rnekleme oranÄ±
    featureSubsetStrategy='sqrt',  # Her bÃ¶lÃ¼nmede sqrt(n) Ã¶zellik
    impurity='gini',            # Gini impurity
    seed=42,
    cacheNodeIds=True,          # Node ID cache (hÄ±z iÃ§in)
    checkpointInterval=10       # Checkpoint aralÄ±ÄŸÄ±
)

# Model parametrelerini gÃ¶ster
print("ðŸ“‹ Model Parametreleri (Bellek Dostu):")
print(f"   - AÄŸaÃ§ SayÄ±sÄ± (numTrees): 50")
print(f"   - Max Derinlik (maxDepth): 8")
print(f"   - Max Bins: 64")
print(f"   - Min Instances Per Node: 20")
print(f"   - Subsampling Rate: 0.7")

# EÄŸitim zamanÄ±nÄ± Ã¶lÃ§
start_time = time.time()

# BÃ¼yÃ¼k veri iÃ§in kademeli eÄŸitim dene
sample_fractions = [1.0, 0.5, 0.3]  # Tam veri, yoksa %50, yoksa %30
rf_model = None
used_fraction = None

for frac in sample_fractions:
    try:
        if frac < 1.0:
            print(f"\nâš ï¸ Bellek yetersiz, %{int(frac*100)} Ã¶rneklem deneniyor...")
            train_sample = train_rf.sample(False, frac, seed=42)
            train_sample = train_sample.persist(StorageLevel.MEMORY_AND_DISK)
            sample_count = train_sample.count()
            print(f"   Ã–rneklem boyutu: {sample_count:,}")
        else:
            print("\nðŸš€ Tam veri ile eÄŸitim deneniyor...")
            train_sample = train_rf
        
        rf_model = rf_classifier.fit(train_sample)
        used_fraction = frac
        print(f"\nâœ… EÄŸitim baÅŸarÄ±lÄ±! (KullanÄ±lan oran: %{int(frac*100)})")
        break
        
    except Exception as e:
        print(f"\nâŒ EÄŸitim hatasÄ± (oran %{int(frac*100)}): {str(e)[:100]}...")
        if frac < 1.0:
            train_sample.unpersist()
        continue

if rf_model is None:
    raise RuntimeError("Random Forest eÄŸitilemedi! Spark bellek ayarlarÄ±nÄ± kontrol edin.")

training_time = time.time() - start_time
print(f"â±ï¸ EÄŸitim sÃ¼resi: {training_time:.2f} saniye")
```

```python
# ============================================================================
# MODEL DEÄžERLENDÄ°RME
# ============================================================================

print("ðŸ“Š Model Performans Metrikleri")
print("="*60)

# Test seti Ã¼zerinde tahmin yap
rf_predictions = rf_model.transform(test_rf)
rf_predictions = rf_predictions.persist(StorageLevel.MEMORY_AND_DISK)

# Evaluator oluÅŸtur
evaluator = MulticlassClassificationEvaluator(
    labelCol='Severity',
    predictionCol='prediction'
)

# Metrikleri hesapla
rf_accuracy = evaluator.evaluate(rf_predictions, {evaluator.metricName: 'accuracy'})
rf_f1 = evaluator.evaluate(rf_predictions, {evaluator.metricName: 'f1'})
rf_precision = evaluator.evaluate(rf_predictions, {evaluator.metricName: 'weightedPrecision'})
rf_recall = evaluator.evaluate(rf_predictions, {evaluator.metricName: 'weightedRecall'})

print(f"\nðŸŽ¯ SONUÃ‡LAR:")
print(f"   â”œâ”€â”€ Accuracy:  {rf_accuracy:.4f} ({rf_accuracy*100:.2f}%)")
print(f"   â”œâ”€â”€ F1-Score:  {rf_f1:.4f}")
print(f"   â”œâ”€â”€ Precision: {rf_precision:.4f}")
print(f"   â””â”€â”€ Recall:    {rf_recall:.4f}")

# Her sÄ±nÄ±f iÃ§in tahmin daÄŸÄ±lÄ±mÄ±
print("\nðŸ“Š Tahmin DaÄŸÄ±lÄ±mÄ±:")
rf_predictions.groupBy('prediction').count().orderBy('prediction').show()
```

```python
# kNN Modeli EgÌ†itimi
print("ðŸ¤– kNN Modeli Egitiliyor...")
print("="*60)

# On kosul: RF egÌ†itimi tamamlanmis olmali
if 'rf_metrics' not in globals():
    raise RuntimeError("Random Forest huÌˆcresi cÌ§alÄ±sÌ§tÄ±rÄ±lmadan kNN egÌ†itimine gecÌ§meyin.")

# FarklÄ± k degerleri icin test
k_values = [3, 5, 7, 9, 11]
results = []

for k in k_values:
    print(f"\nðŸ¤– k={k} icin model egÌ†itiliyor...")

    knn = SklearnKNN(n_neighbors=k, weights='distance', n_jobs=-1)
    knn.fit(X_train_scaled, y_train)

    # Tahmin
    y_pred = knn.predict(X_test_scaled)

    # Accuracy hesapla
    accuracy = np.mean(y_pred == y_test)
    results.append({'k': k, 'accuracy': accuracy})

    print(f"   Accuracy: {accuracy:.4f}")

# En iyi k degerini bul
best_result = max(results, key=lambda x: x['accuracy'])
print(f"\nðŸ¤– En iyi k degeri: {best_result['k']} (Accuracy: {best_result['accuracy']:.4f})")
```

## ðŸŽ¨ BÃ¶lÃ¼m 7: K-Means KÃ¼meleme + GÃ¶rselleÅŸtirme


```python
# K-Means iÃ§in Ã¶zellik seÃ§imi
# CoÄŸrafi ve Ã§evresel faktÃ¶rlere gÃ¶re kazalarÄ± kÃ¼meleyeceÄŸiz

KMEANS_TARGET_ROWS = 2_000_000
KMEANS_ELBOW_SAMPLE = 200_000

kmeans_total_rows = total_count if 'total_count' in globals() else df_sample.count()
kmeans_fraction = min(1.0, KMEANS_TARGET_ROWS / kmeans_total_rows)

# Rastgele Ã¶rnekle ve Ã¼st sÄ±nÄ±rÄ± sabitle
df_kmeans_base = df_sample.sample(withReplacement=False, fraction=kmeans_fraction, seed=42)
df_kmeans_base = df_kmeans_base.limit(KMEANS_TARGET_ROWS)
df_kmeans_base.cache()
kmeans_used_rows = df_kmeans_base.count()

print("ðŸ“Š K-Means iÃ§in Ã¶rnek hazÄ±rlama")
print("="*60)
print(f"   â€¢ Hedef: {KMEANS_TARGET_ROWS:,} satÄ±r")
print(f"   â€¢ Toplam veri: {kmeans_total_rows:,} satÄ±r")
print(f"   â€¢ KullanÄ±lan: {kmeans_used_rows:,} satÄ±r")

kmeans_features = [
    'Start_Lat', 'Start_Lng', 'Temperature(F)', 'Humidity(%)',
    'Visibility(mi)', 'Wind_Speed(mph)'
]

# PySpark VectorAssembler ile Ã¶zellik vektÃ¶rÃ¼ oluÅŸtur
assembler = VectorAssembler(inputCols=kmeans_features, outputCol="features", handleInvalid="skip")
df_kmeans = assembler.transform(df_kmeans_base)

# Ã–lÃ§ekleme
scaler_spark = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=True)
scaler_model = scaler_spark.fit(df_kmeans)
df_kmeans_scaled = scaler_model.transform(df_kmeans)

print("âœ… K-Means iÃ§in Ã¶zellikler hazÄ±rlandÄ±!")
```

```python
# Elbow + Silhouette analizi - Optimal k deÄŸerini bul
print("ðŸ“Š Elbow + Silhouette Analizi")
print("="*60)

# Daha kÃ¼Ã§Ã¼k bir Ã¶rnekle Ã§alÄ±ÅŸ (hÄ±z iÃ§in)
eval_fraction = min(1.0, KMEANS_ELBOW_SAMPLE / max(kmeans_used_rows, 1))
df_kmeans_sample = df_kmeans_scaled.sample(withReplacement=False, fraction=eval_fraction, seed=42)
df_kmeans_sample.cache()
eval_count = df_kmeans_sample.count()

k_range = list(range(2, 11))
costs = []
silhouettes = []

evaluator = ClusteringEvaluator(
    featuresCol="scaled_features",
    predictionCol="cluster",
    metricName="silhouette",
    distanceMeasure="squaredEuclidean"
)

for k in k_range:
    print(f"ðŸ”„ k={k} test ediliyor...")

    kmeans = KMeans(k=k, seed=42, featuresCol="scaled_features", predictionCol="cluster")
    model = kmeans.fit(df_kmeans_sample)

    # WSSSE (Within Set Sum of Squared Errors)
    cost = model.summary.trainingCost
    costs.append(cost)

    # Silhouette
    sil = evaluator.evaluate(model.transform(df_kmeans_sample))
    silhouettes.append(sil)

    print(f"   WSSSE: {cost:.2f} | Silhouette: {sil:.4f}")

# En iyi k (silhouette'e gÃ¶re)
best_idx = int(np.argmax(silhouettes))
OPTIMAL_K = k_range[best_idx]
print(f"\nðŸ† En iyi k (Silhouette): {OPTIMAL_K} | Skor: {silhouettes[best_idx]:.4f}")
```

```python
# Elbow grafiÄŸi + Silhouette eÄŸrisi
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(k_range, costs, 'bo-', linewidth=2, markersize=10, label='WSSSE')
ax1.set_xlabel('KÃ¼me SayÄ±sÄ± (k)')
ax1.set_ylabel('WSSSE (Within Set Sum of Squared Errors)')
ax1.grid(True, alpha=0.3)

ax2 = ax1.twinx()
ax2.plot(k_range, silhouettes, 'gs--', linewidth=2, markersize=8, label='Silhouette')
ax2.set_ylabel('Silhouette Skoru')

ax1.axvline(x=OPTIMAL_K, color='red', linestyle='--', label=f'Optimal k={OPTIMAL_K}')

# Legend birleÅŸtir
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='best')

plt.title('K-Means: Elbow + Silhouette Analizi')
plt.tight_layout()
plt.savefig('elbow_method.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Final K-Means Modeli
print(f"ðŸŽ¯ Final K-Means Modeli (k={OPTIMAL_K})")
print("="*60)

kmeans_final = KMeans(k=OPTIMAL_K, seed=42, featuresCol="scaled_features", predictionCol="cluster")
kmeans_model = kmeans_final.fit(df_kmeans_scaled)

# KÃ¼me atamalarÄ±
df_clustered = kmeans_model.transform(df_kmeans_scaled)

# KÃ¼me merkezleri
centers = kmeans_model.clusterCenters()
print("\nðŸ“ KÃ¼me Merkezleri:")
for i, center in enumerate(centers):
    print(f"   KÃ¼me {i}: {center[:3]}... (ilk 3 Ã¶zellik)")
```

```python
# KÃ¼me daÄŸÄ±lÄ±mÄ±
print("\nðŸ“Š KÃ¼me DaÄŸÄ±lÄ±mÄ±:")
cluster_dist = df_clustered.groupBy('cluster').count().orderBy('cluster')
cluster_dist.show()

cluster_pd = cluster_dist.toPandas()

plt.figure(figsize=(10, 6))
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7']
plt.bar(cluster_pd['cluster'].astype(str), cluster_pd['count'], color=colors[:len(cluster_pd)])
plt.xlabel('KÃ¼me')
plt.ylabel('KayÄ±t SayÄ±sÄ±')
plt.title('K-Means KÃ¼me DaÄŸÄ±lÄ±mÄ±')
for i, v in enumerate(cluster_pd['count']):
    plt.text(i, v + 500, f'{v:,}', ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('cluster_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```
